```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta, binom

plt.style.use('ggplot')
```

### 1. Introduction 

This notebook will focus on the beta distribution, which plays a fundamental role in Bayesian statistics.  A beta distribution has two parameters that control it, $\alpha$ and $\beta$, and takes a range of shapes on the interval [0,1]. 

```python
# Beta distribution function
def plot_beta(x_range, a, b, mu=0, sigma=1, cdf=False, **kwargs):
    x = x_range
    if cdf:
        y = beta.cdf(x, a, b, mu, sigma)
    else: 
        y = beta.pdf(x, a, b, mu, sigma)
    plt.plot(x, y, **kwargs)
```

```python
x = np.linspace(0, 1, 5000)
plot_beta(x, 1, 2, 0, 1, color='red', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 1, $\beta$ = 2')
plot_beta(x, 3, 3, 0, 1, color='green', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 3, $\beta$ = 3')
plot_beta(x, 20, 20, 0, 1, color='blue', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 20, $\beta$ = 20')
plot_beta(x, 50, 10, 0, 1, color='purple', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 50, $\beta$ = 10')
plt.xlabel('Batting Average')
plt.ylabel('Density of Beta')
plt.legend()
plt.show()
```

The beta distribution represents a probability distribution of *probabilities*.  If we don't know what the probability of something is, the beta distribution represents well a distribution of probabilities for that value.  

### 2. Batting Averages

The batting average in baseball is defined as: 

Batting Average = $\frac{H}{AB}$

where H represents number of hits and AB represents number of at-bats.  If we want to think about a player's long-term batting average, looking at only a few hits will not give us a good indication of what the *true* batting average is.  For example, if the season just started and we see that a player had 4 hits at 4 at-bats, we wouldn't seriously believe his long-term batting average will remain around 1, when the league average is below 0.300.  

The number of hits a player gets out of the at-bats is modeled by a binomial distribution, which calculates the number of successes out of a total number of independent trials.  Since the binomial distribution has the beta distribution as a conjugate prior, we can represent the prior expectations for hits with a beta distribution.  So this distribution represents, before seeing any at-bats, what we expect the batting average for the player to be.  We expect the player's season-long average to be most likely around .27, but could be anywhere from .21 to .35.  We can represent this with a beta distribution with parameters $\alpha$ = 81 and $\beta$ = 219: 

```python
# Beta distribution with parameters alpha=81 and beta=219
x = np.linspace(0, 1, 5000)
plot_beta(x, 81, 219, 0, 1, color='red', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 81, $\beta$ = 219')
plt.xlabel('Batting Average')
plt.ylabel('Density of Beta')
plt.legend()
plt.show()
```

The great thing about Bayesian analysis is that anytime a player makes another hit and has another at-bat, we can just shift our distribution! This is as simple as making the new beta distribution: Beta($\alpha_{0}$+ hits, $\beta_{0}$ + misses), so our new beta distribution is just Beta(81 + 1, 219)

```python
# New beta distribution after recording one hit 
x = np.linspace(0, 1, 5000)
plot_beta(x, 81, 219, 0, 1, color='red', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 81, $\beta$ = 219')
plot_beta(x, 82, 219, 0, 1, color='green', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 82, $\beta$ = 219')
plot_beta(x, 181, 419, 0, 1, color='blue', lw=2, ls='-', alpha=0.7, label=r'$\alpha$ = 181, $\beta$ = 419')
plt.xlabel('Batting Average')
plt.ylabel('Density of Beta')
plt.legend()
plt.show()
```

The above plot shows how the beta distribution would change in response to one hit, and then the blue distribution represents our update half-way through the season with 100 hits and 200 misses.  The blue distribution is both thinner and shifted to the right, indicating we have more certainty about what the player's batting average is.  

### 3. Posterior Mean 

The expected value, or mean, of the beta distribution is given by:

$\frac{\alpha}{\alpha + \beta}$

So after 100 hits out of 300 at-bats, the expected value of the new distribution is: 

$\frac{82 + 100}{82 + 100 + 219 + 200}$ = .303

So this is lower than our raw estimate of .333, but higher than the estimate we started with, .270.  So this is a combination of our prior expectations and estimates! This is the real power of Bayesian analysis.  

### 4. Conjugate Prior 

Why is it so easy to update the beta distribution? The reason is because the beta is the *conjugate prior* of the binomial distribution.  Suppose that we simulate 10,000,000 baseball players, who all bat according to our prior distribution, $\beta$(81, 219).  We will give each of them 300 chances at the plate, just like our 100/300 player: 

```python
# Simulate 10,000,000 baseball players from beta(81, 219)
num_trials = 10000000
simulations = pd.DataFrame({'true_average': np.random.beta(81, 219, size=num_trials)})
simulations['hits'] = np.random.binomial(300, simulations['true_average'], num_trials)
simulations.head()
```

How many of our hypothetical players got 100/300 hits? Let's check: 

```python
# Players who got 100/300 hits
hit_100 = simulations[simulations['hits'] == 100]
hit_100.shape
```

So it looks like there were 80,226 players who got exactly 100 hits out of our simulation.  What was the distribution of batting averages for the 100/300 players? How good was the median player? Let's also check: 

```python
# Histogram of batting averages for 100 hit players
sns.distplot(hit_100['true_average'], hist_kws=dict(edgecolor="g", linewidth=1));
```

In our prior distribution, there were batters with a true batting average of .2, but they never achieve 100/300 in the simulation.  And even for people with a .330 batting average, there aren't many who get 100/300 hits.  The median player who got 100/300 has a batting average of about .3, which is our posterior estimate.  Our plot shows that Bayesian updating is asking "out of our prior, what kinds of players would end up with evidence like this?"
